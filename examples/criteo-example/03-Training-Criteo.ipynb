# Training with Merlin Models

```python
import os
import pandas as pd
import nvtabular as nvt
from merlin.models.utils.example_utils import workflow_fit_transform
import merlin.io

import merlin.models.tf as mm

from nvtabular import ops
from merlin.core.utils import download_file
from merlin.datasets.entertainment import get_movielens
from merlin.schema.tags import Tags
```

```python
input_path="/data/test_dask/output"

train = merlin.io.Dataset(
    os.path.join(input_path, "train"), engine="parquet"
)
valid = merlin.io.Dataset(
    os.path.join(input_path, "valid"), engine="parquet"
)
```

```python
train.schema
```

```python
model = mm.DLRMModel(
    train.schema,
    embedding_dim=32, # From HugeCTR
    bottom_block=mm.MLPBlock([128, 32]),
    top_block=mm.MLPBlock([128, 64, 32]),
    prediction_tasks=mm.BinaryClassificationTask(
        train.schema.select_by_tag(Tags.TARGET).column_names[0]
    ),
)

model.compile(optimizer="adam")
model.fit(train, batch_size=128)
```

```python
metrics = model.evaluate(valid, batch_size=128, return_dict=True)
```
